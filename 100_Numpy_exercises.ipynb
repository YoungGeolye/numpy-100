#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)

````python
A=np.ones(3)*1
B=np.ones(3)*2
print(A)
print(B)
np.add(A,B,out=B)
np.negative(A,out=A)
np.divide(A,2,out=A)
np.multiply(A,B,out=A)
print(A)
````

#### 44. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

````python
Z=np.random.random((10,2))
X,Y=Z[:,0],Z[:,1]
R=np.sqrt(X**2+Y**2)
T=np.arctan2(Y,X)
print(R)
print(T)
polar_coord=np.column_stack((R,T))
print(polar_coord)
polar_coord=np.stack((R,T),axis=1)
print(polar_coord)
````
